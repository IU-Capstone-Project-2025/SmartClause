```python
import math
import json
from collections import Counter, defaultdict
from typing import List, Dict, Sequence, Tuple

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sqlalchemy.engine import Engine
from sentence_transformers import SentenceTransformer

# -----------------------------------------------------------------------------
# TEXT PRE‑/POST‑PROCESS HELPERS
# -----------------------------------------------------------------------------

def _tokenise(text: str) -> List[str]:
    """Lower‑case split and keep alphabetic tokens only."""
    return [w for w in text.lower().split() if w.isalpha()]


def preprocess_text_for_metrics(text) -> set:
    """Safely preprocess *any* text for metric calculations.

    Returns an empty set for ``None`` or empty strings, so downstream code never
    hits the dreaded ``NoneType has no attribute 'lower'`` error.
    """
    if not text:
        return set()
    return set(_tokenise(text))

# -----------------------------------------------------------------------------
# NAÏVE BM25 (no external dependency)
# -----------------------------------------------------------------------------

def _simple_bm25(
    corpus_tokens: List[List[str]],
    query_tokens: List[str],
    k1: float = 1.5,
    b: float = 0.75,
) -> np.ndarray:
    """Return BM25 *similarity* scores for every document in the corpus."""
    N = len(corpus_tokens)
    if N == 0:
        return np.zeros(0)

    avgdl = sum(len(d) for d in corpus_tokens) / N

    df = defaultdict(int)
    for doc in corpus_tokens:
        for w in set(doc):
            df[w] += 1
    idf = {w: math.log1p((N - f + 0.5) / (f + 0.5)) for w, f in df.items()}

    scores = np.zeros(N)
    for i, doc in enumerate(corpus_tokens):
        freq = Counter(doc)
        dl = len(doc)
        for w in query_tokens:
            if w not in freq:
                continue
            tf = freq[w]
            scores[i] += idf.get(w, 0.0) * (tf * (k1 + 1)) / (tf + k1 * (1 - b + b * dl / avgdl))
    return scores


# -----------------------------------------------------------------------------
# VECTOR DISTANCE UTILITIES
# -----------------------------------------------------------------------------

def _vector_similarity(vec_a: List[float], vec_b: List[float], metric: str = "cosine") -> float:
    if metric == "inner_product":
        return float(np.dot(vec_a, vec_b))
    elif metric == "euclidean":
        return -float(np.linalg.norm(np.subtract(vec_a, vec_b)))  # negative dist == similarity
    # default cosine
    denom = math.sqrt(sum(a * a for a in vec_a)) * math.sqrt(sum(b * b for b in vec_b))
    return float(np.dot(vec_a, vec_b) / denom) if denom else 0.0


# -----------------------------------------------------------------------------
# RETRIEVAL: PGVECTOR (single metric)
# -----------------------------------------------------------------------------

def retrieve_articles_for_document(
    doc_text: str,
    k: int = 5,
    engine: Engine | None = None,
    distance_metric: str = "cosine",
) -> List[str]:
    """Top‑k rule texts via pgvector operator decided by *distance_metric*."""

    model = SentenceTransformer("BAAI/bge-m3")
    query_emb = model.encode(doc_text).tolist()
    vec_literal = f"[{','.join(map(str, query_emb))}]"
    op = {"cosine": f"(1 - (rc.embedding <=> '{vec_literal}'))", 
          "inner_product": f"((rc.embedding <#> '{vec_literal}') * -1)", 
          "euclidean": f"(rc.embedding <-> '{vec_literal}')"}[distance_metric]

    sql = f"""
        WITH best AS (
            SELECT DISTINCT ON (r.rule_id)
                r.rule_id,
                r.rule_text,
                {op} AS sim
            FROM rule_chunks rc
            JOIN rules r ON r.rule_id = rc.rule_id
            ORDER BY r.rule_id, sim DESC
        )
        SELECT rule_text
        FROM best
        ORDER BY sim DESC
        LIMIT {k};
    """
    df = pd.read_sql(sql, engine)
    return df["rule_text"].dropna().tolist()


# -----------------------------------------------------------------------------
# RETRIEVAL: BM25 + VECTOR + RRF (multi‑metric)
# -----------------------------------------------------------------------------

def retrieve_articles_rrf(
    doc_text: str,
    k: int = 5,
    engine: Engine | None = None,
    distance_metric: str = "cosine",
    c: int = 60,
) -> List[str]:
    """Hybrid retrieval with BM25 & vector similarity fused via RRF."""
    # 1) fetch chunks
    df = pd.read_sql(
        """
        SELECT rc.chunk_id, rc.chunk_text, rc.embedding, r.rule_id
        FROM rule_chunks rc JOIN rules r ON r.rule_id = rc.rule_id
        WHERE rc.embedding IS NOT NULL
        """,
        engine,
    )
    if df.empty:
        return []

    # 2) pre‑compute lists
    texts = df["chunk_text"].tolist()
    corpus_tokens = [_tokenise(t) for t in texts]
    q_tokens = _tokenise(doc_text)

    # 3) BM25 scores & ranks
    bm25_scores = _simple_bm25(corpus_tokens, q_tokens)
    bm25_rank = np.argsort(-bm25_scores).tolist()

    # 4) Vector scores & ranks
    model = SentenceTransformer("BAAI/bge-m3")
    q_emb = model.encode(doc_text).tolist()

    vector_scores = [
        _vector_similarity(json.loads(e) if isinstance(e, str) else e, q_emb, metric=distance_metric)
        for e in df["embedding"]
    ]
    vector_rank = np.argsort(-np.array(vector_scores)).tolist()

    # 5) RRF fusion
    def _rrf_idx(idx: int, ranking: List[int]) -> int:
        try:
            return ranking.index(idx)
        except ValueError:
            return len(ranking)

    scores = {
        i: 1 / (c + _rrf_idx(i, bm25_rank)) + 1 / (c + _rrf_idx(i, vector_rank))
        for i in range(len(df))
    }

    fused_rank = sorted(scores.keys(), key=lambda i: scores[i], reverse=True)

    # 6) collect *unique rule* texts until k reached
    seen: set[int] = set()
    result = []
    for idx in fused_rank:
        rule_id = int(df.at[idx, "rule_id"])
        if rule_id in seen:
            continue
        seen.add(rule_id)
        text_sql = "SELECT rule_text FROM rules WHERE rule_id = %s LIMIT 1"
        rule_df = pd.read_sql(text_sql, engine, params=(rule_id,))
        if not rule_df.empty and (rtxt := rule_df.iloc[0][0]):
            result.append(rtxt)
        if len(result) >= k:
            break
    return result


# -----------------------------------------------------------------------------
# QUALITY METRICS
# -----------------------------------------------------------------------------

def _calculate_relevance(chunk_words: set, reference_words: set) -> float:
    if not chunk_words or not reference_words:
        return 0.0
    inter = chunk_words & reference_words
    return len(inter) / min(len(chunk_words), len(reference_words))


def _average_relevance(chunks: List[set], reference: set) -> float:
    if not chunks:
        return 0.0
    return sum(_calculate_relevance(c, reference) for c in chunks) / len(chunks)


def _context_completeness(retrieved: set, reference: set) -> float:
    return len(retrieved & reference) / len(reference) if reference else 0.0


def _context_relevance(retrieved: set, reference: set) -> float:
    return len(retrieved & reference) / len(retrieved) if retrieved else 0.0


# -----------------------------------------------------------------------------
# EVALUATORS (vector & rrf) – both return Dict[str|metric, DataFrame]
# -----------------------------------------------------------------------------

def evaluate_vector(
    docs: Sequence[str],
    references: Sequence[Sequence[str]],
    engine: Engine,
    k: int = 5,
    distance_metrics: Sequence[str] = ("cosine", "inner_product", "euclidean"),
) -> Dict[str, pd.DataFrame]:
    out = {}
    print("Evaluation of simple retrieval")
    for metric in distance_metrics:
        print(metric)
        rows = []
        for i, (doc, ref_texts) in enumerate(zip(docs, references), start=1):
            retrieved = retrieve_articles_for_document(doc, k=k, engine=engine, distance_metric=metric)
            ref_words = preprocess_text_for_metrics(" ".join(t for t in ref_texts if t))
            ret_words = preprocess_text_for_metrics(" ".join(retrieved))
            chunk_sets = [preprocess_text_for_metrics(a) for a in retrieved]
            inter_cnt = _intersection_count(retrieved, ref_texts)
            rows.append(
                {
                    "doc_id": i,
                    "retrieved": len(retrieved),
                    "reference": len(ref_texts),
                    "intersection": inter_cnt,
                    "avg_rel": _average_relevance(chunk_sets, ref_words),
                    "ctx_comp": _context_completeness(ret_words, ref_words),
                    "ctx_rel": _context_relevance(ret_words, ref_words),
                }
            )
            print(i)
        out[metric] = pd.DataFrame(rows)
    return out


def evaluate_rrf(
    docs: Sequence[str],
    references: Sequence[Sequence[str]],
    engine: Engine,
    k: int = 5,
    distance_metrics: Sequence[str] = ("cosine", "inner_product", "euclidean"),
) -> Dict[str, pd.DataFrame]:
    out = {}
    print("Evaluation of rrf retrieval")
    for metric in distance_metrics:
        print(metric)
        rows = []
        for i, (doc, ref_texts) in enumerate(zip(docs, references), start=1):
            retrieved = retrieve_articles_rrf(doc, k=k, engine=engine, distance_metric=metric)
            ref_words = preprocess_text_for_metrics(" ".join(t for t in ref_texts if t))
            ret_words = preprocess_text_for_metrics(" ".join(retrieved))
            chunk_sets = [preprocess_text_for_metrics(a) for a in retrieved]
            inter_cnt = _intersection_count(retrieved, ref_texts)
            rows.append(
                {
                    "doc_id": i,
                    "retrieved": len(retrieved),
                    "reference": len(ref_texts),
                    "intersection": inter_cnt,
                    "avg_rel": _average_relevance(chunk_sets, ref_words),
                    "ctx_comp": _context_completeness(ret_words, ref_words),
                    "ctx_rel": _context_relevance(ret_words, ref_words),
                }
            )
            print(i)
        out[metric] = pd.DataFrame(rows)
    return out


# ------------------------- INTERNAL HELPERS ----------------------------------

def _intersection_count(retrieved: List[str], reference_list: Sequence[str]) -> int:
    cnt, used = 0, set()
    for art in retrieved:
        art_w = preprocess_text_for_metrics(art)
        for idx, ref in enumerate(reference_list):
            if idx in used:
                continue
            ref_w = preprocess_text_for_metrics(ref)
            if ref_w and len(art_w & ref_w) / len(ref_w) > 0.3:
                cnt += 1
                used.add(idx)
                break
    return cnt


# -----------------------------------------------------------------------------
# VISUALISATION UTILITIES (one DF → three plots)
# -----------------------------------------------------------------------------

def _plot_intersections(df: pd.DataFrame, title: str):
    x = range(len(df))
    width = 0.35
    plt.figure(figsize=(14, 6))
    plt.bar([i - width/2 for i in x], df["reference"], width, label="Reference", alpha=0.7)
    plt.bar([i + width/2 for i in x], df["intersection"], width, label="Intersection", alpha=0.7)
    plt.xticks(x, [f"Doc {d}" for d in df["doc_id"]])
    plt.ylabel("Count")
    plt.title(title)
    plt.legend()
    plt.grid(alpha=0.3)
    plt.tight_layout()
    plt.show()


def _plot_metric_hists(df: pd.DataFrame, title_prefix: str):
    metrics = {"avg_rel": "Average Relevance", "ctx_comp": "Context Completeness", "ctx_rel": "Context Relevance"}
    plt.figure(figsize=(15, 4))
    for i, (col, lab) in enumerate(metrics.items(), 1):
        plt.subplot(1, 3, i)
        plt.hist(df[col], bins=10, edgecolor="black", alpha=0.7)
        plt.axvline(df[col].mean(), linestyle="--")
        plt.title(f"{title_prefix} – {lab}")
        plt.grid(alpha=0.3)
    plt.tight_layout()
    plt.show()


def _plot_box(df: pd.DataFrame, title: str):
    melt = df[["avg_rel", "ctx_comp", "ctx_rel"]].melt(var_name="Metric", value_name="Value")
    plt.figure(figsize=(8, 6))
    sns.boxplot(data=melt, x="Metric", y="Value", palette="Set3")
    sns.stripplot(data=melt, x="Metric", y="Value", color="red", alpha=0.5)
    plt.title(title)
    plt.grid(alpha=0.3)
    plt.tight_layout()
    plt.show()


# -----------------------------------------------------------------------------
# PUBLIC WRAPPERS
# -----------------------------------------------------------------------------

def visualize_vector(df_by_metric: Dict[str, pd.DataFrame]):
    for m, df in df_by_metric.items():
        _plot_intersections(df, f"Vector ({m}) – Intersections")
        _plot_metric_hists(df, f"Vector ({m})")
        _plot_box(df, f"Vector ({m}) – Metric Distribution")


def visualize_rrf(df_by_metric: Dict[str, pd.DataFrame]):
    for m, df in df_by_metric.items():
        _plot_intersections(df, f"RRF ({m}) – Intersections")
        _plot_metric_hists(df, f"RRF ({m})")
        _plot_box(df, f"RRF ({m}) – Metric Distribution")


def compare_means(df_base: pd.DataFrame, df_rrf: pd.DataFrame, label_base: str = "Vector", label_rrf: str = "RRF"):
    means_base = df_base[["avg_rel", "ctx_comp", "ctx_rel"]].mean()
    means_rrf = df_rrf[["avg_rel", "ctx_comp", "ctx_rel"]].mean()
    idx = np.arange(len(means_base))
    width = 0.35
    plt.figure(figsize=(8, 5))
    plt.bar(idx - width/2, means_base, width, label=label_base)
    plt.bar(idx + width/2, means_rrf, width, label=label_rrf)
    plt.xticks(idx, ["AvgRel", "CtxComp", "CtxRel"])
    plt.ylabel("Mean score")
    plt.title("Mean metric comparison")
    plt.legend()
    plt.grid(alpha=0.3)
    plt.tight_layout()
    plt.show()
    improv = (means_rrf - means_base) / means_base * 100
    print(improv.to_frame("% Improvement").to_string(float_format="%+.1f%%"))

```

```python
vec_results = evaluate_vector(
    docs=filtered_docs,
    references=filtered_proof_texts,
    engine=engine,
    k=5,
    distance_metrics=("cosine", "inner_product", "euclidean")
)

rrf_results = evaluate_rrf(
    docs=filtered_docs,
    references=filtered_proof_texts,
    engine=engine,
    k=5,
    distance_metrics=("cosine", "inner_product", "euclidean")
)


visualize_vector(vec_results) 
visualize_rrf(rrf_results)    

compare_means(
    vec_results["cosine"],   
    rrf_results["cosine"],    
    label_base="Vector-Cosine",
    label_rrf="RRF-Cosine")
```
