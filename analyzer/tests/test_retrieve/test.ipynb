```python
def retrieve_articles_for_document(doc_text, k=5, engine=None):
    """
    Упрощенная функция ретрива статей для документа.
    Повторяет пайплайн из retrieval_service без импорта дополнительных классов.
    
    Args:
        doc_text (str): Текст документа для поиска
        k (int): Количество статей для возврата
        engine: SQLAlchemy engine для подключения к БД
    
    Returns:
        list: Список текстов найденных статей
    """
    import json
    import numpy as np
    from sentence_transformers import SentenceTransformer
    
    # 1. Генерация эмбеддинга для документа
    try:
        # Используем ту же модель, что и в сервисе
        model = SentenceTransformer('BAAI/bge-m3')
        query_embedding = model.encode(doc_text).tolist()
    except Exception as e:
        print(f"Ошибка генерации эмбеддинга: {e}")
        return []
    
    # 2. Конвертация в формат PostgreSQL
    query_vector = f"[{','.join(map(str, query_embedding))}]"
    
    # 3. SQL-запрос для поиска по косинусному расстоянию
    sql_query = f"""
        WITH rule_best_chunks AS (
            SELECT DISTINCT ON (r.rule_id)
                r.rule_id,
                r.file,
                r.rule_number,
                r.rule_title,
                r.rule_text,
                r.section_title,
                r.chapter_title,
                rc.embedding,
                (1 - (rc.embedding <=> '{query_vector}')) as similarity_score
            FROM rule_chunks rc
            JOIN rules r ON rc.rule_id = r.rule_id
            WHERE rc.embedding IS NOT NULL
            ORDER BY r.rule_id, similarity_score DESC
        )
        SELECT 
            rule_text,
            file,
            rule_number,
            rule_title,
            similarity_score
        FROM rule_best_chunks
        ORDER BY similarity_score DESC
        LIMIT {k}
    """
    
    # 4. Выполнение запроса
    try:
        df = pd.read_sql(sql_query, engine)
        if df.empty:
            print("Не найдено статей для документа")
            return []
        
        # 5. Возврат текстов статей
        articles = []
        for _, row in df.iterrows():
            if pd.notna(row['rule_text']):
                articles.append(row['rule_text'])
        
        print(f"Найдено {len(articles)} статей для документа")
        return articles
        
    except Exception as e:
        print(f"Ошибка выполнения запроса: {e}")
        return []

```

```python
def preprocess_text_for_metrics(text):
    """
    Предобработка текста для расчета метрик.
    Убирает пунктуацию, приводит к нижнему регистру, токенизирует.
    """
    import re
    if not text:
        return set()
    
    # Убираем пунктуацию и приводим к нижнему регистру
    text_clean = re.sub(r'[^\w\s]', ' ', text.lower())
    # Разбиваем на слова и убираем пустые
    words = [word.strip() for word in text_clean.split() if word.strip()]
    return set(words)

def calculate_relevance(chunk_words, reference_words):
    """
    Рассчитывает Relevance для одного чанка.
    
    Args:
        chunk_words (set): Множество слов в чанке
        reference_words (set): Множество "золотых" слов (эталон)
    
    Returns:
        float: Значение Relevance
    """
    if not chunk_words or not reference_words:
        return 0.0
    
    intersection = chunk_words & reference_words
    min_length = min(len(chunk_words), len(reference_words))
    
    if min_length == 0:
        return 0.0
    
    return len(intersection) / min_length

def calculate_average_relevance(chunks_words, reference_words):
    """
    Рассчитывает Average Relevance по всем чанкам.
    
    Args:
        chunks_words (list): Список множеств слов для каждого чанка
        reference_words (set): Множество "золотых" слов (эталон)
    
    Returns:
        float: Значение Average Relevance
    """
    if not chunks_words:
        return 0.0
    
    relevances = [calculate_relevance(chunk_words, reference_words) for chunk_words in chunks_words]
    return sum(relevances) / len(relevances)

def calculate_context_completeness(retrieved_words, reference_words):
    """
    Рассчитывает Context Completeness (CC).
    
    Args:
        retrieved_words (set): Объединенное множество слов всех retrieved статей
        reference_words (set): Множество "золотых" слов (эталон)
    
    Returns:
        float: Значение Context Completeness
    """
    if not reference_words:
        return 0.0
    
    intersection = retrieved_words & reference_words
    return len(intersection) / len(reference_words)

def calculate_context_relevance(retrieved_words, reference_words):
    """
    Рассчитывает Context Relevance (CR).
    
    Args:
        retrieved_words (set): Объединенное множество слов всех retrieved статей
        reference_words (set): Множество "золотых" слов (эталон)
    
    Returns:
        float: Значение Context Relevance
    """
    if not retrieved_words:
        return 0.0
    
    intersection = retrieved_words & reference_words
    return len(intersection) / len(retrieved_words)
```

```python
def evaluate_retrieval_metrics_with_intersections(filtered_docs, filtered_proof_texts, k=5):
    """
    Оценивает качество ретрива с правильным подсчетом пересечений.
    """
    results = []
    
    for i, (doc, reference_texts_list) in enumerate(zip(filtered_docs, filtered_proof_texts)):
        print(f"\n--- Оценка документа {i+1} ---")
        
        # 1. Получаем статьи через ретрив
        retrieved_articles = retrieve_articles_for_document(doc, k=k, engine=engine)
        
        if not retrieved_articles:
            print("Не найдено статей через ретрив, пропускаем документ")
            continue
        
        # 2. Подготавливаем эталонные тексты
        reference_text_combined = " ".join([text for text in reference_texts_list if text])
        reference_words = preprocess_text_for_metrics(reference_text_combined)
        
        # 3. Подготавливаем retrieved тексты
        retrieved_text_combined = " ".join(retrieved_articles)
        retrieved_words = preprocess_text_for_metrics(retrieved_text_combined)
        
        # 4. Подготавливаем отдельные чанки для Average Relevance
        chunks_words = [preprocess_text_for_metrics(article) for article in retrieved_articles]
        
        # 5. Рассчитываем метрики
        avg_relevance = calculate_average_relevance(chunks_words, reference_words)
        context_completeness = calculate_context_completeness(retrieved_words, reference_words)
        context_relevance = calculate_context_relevance(retrieved_words, reference_words)
        
        # 6. Правильно считаем пересечения
        intersection_count = 0
        found_reference_indices = set()  # Чтобы не считать одну эталонную статью дважды
        
        for retrieved_article in retrieved_articles:
            retrieved_article_words = preprocess_text_for_metrics(retrieved_article)
            
            for ref_idx, reference_text in enumerate(reference_texts_list):
                if reference_text and ref_idx not in found_reference_indices:
                    reference_text_words = preprocess_text_for_metrics(reference_text)
                    # Если есть значительное пересечение слов, считаем что статья найдена
                    if len(retrieved_article_words & reference_text_words) / len(reference_text_words) > 0.3:
                        intersection_count += 1
                        found_reference_indices.add(ref_idx)
                        break  # Нашли эту эталонную статью, переходим к следующей найденной
        
        # 7. Сохраняем результаты
        doc_result = {
            'doc_id': i+1,
            'retrieved_articles_count': len(retrieved_articles),
            'reference_articles_count': len([text for text in reference_texts_list if text]),
            'intersection_count': intersection_count,  # Количество пересечений (не больше эталонных)
            'reference_words_count': len(reference_words),
            'retrieved_words_count': len(retrieved_words),
            'average_relevance': avg_relevance,
            'context_completeness': context_completeness,
            'context_relevance': context_relevance
        }
        
        results.append(doc_result)
        
        print(f"Эталонных статей: {doc_result['reference_articles_count']}")
        print(f"Статей найдено: {len(retrieved_articles)}")
        print(f"Пересечений найдено: {intersection_count}")
        print(f"Average Relevance: {avg_relevance:.4f}")
        print(f"Context Completeness: {context_completeness:.4f}")
        print(f"Context Relevance: {context_relevance:.4f}")
    
    return results

# Запускаем оценку с пересечениями
metrics_results_with_intersections = evaluate_retrieval_metrics_with_intersections(filtered_docs, filtered_proof_texts, k=5)
df_metrics_with_intersections = pd.DataFrame(metrics_results_with_intersections)
```

```python
def retrieve_articles_rrf(doc_text, k=5, engine=None):
    """
    Упрощенная функция RRF ретрива статей (BM25 + Vector + RRF).
    Повторяет пайплайн из retrieval_service без импорта дополнительных классов.
    """
    import json
    import numpy as np
    from sentence_transformers import SentenceTransformer
    import math
    from collections import Counter, defaultdict
    
    # 1. Получаем все чанки из базы данных
    query = """
        SELECT rc.chunk_id, rc.chunk_text, rc.embedding, r.rule_id, r.file, r.rule_number, r.rule_title
        FROM rule_chunks rc
        JOIN rules r ON rc.rule_id = r.rule_id
        WHERE rc.embedding IS NOT NULL
        ORDER BY rc.chunk_id
    """
    
    try:
        df_chunks = pd.read_sql(query, engine)
        if df_chunks.empty:
            print("Не найдено чанков в базе данных")
            return []
        
        chunks = df_chunks.to_dict('records')
        texts = [c['chunk_text'] for c in chunks]
        ids = [c['chunk_id'] for c in chunks]
        embeddings = []
        
        # Парсим эмбеддинги
        for chunk in chunks:
            if chunk['embedding']:
                if isinstance(chunk['embedding'], str):
                    emb = json.loads(chunk['embedding'])
                else:
                    emb = chunk['embedding']
                embeddings.append(emb)
            else:
                embeddings.append([0.0] * 1024)  # Заглушка
        
        # 2. BM25 рейтинг
        def preprocess(text):
            tokens = [w for w in text.lower().split() if w.isalpha()]
            return tokens
        
        corpus = [preprocess(text) for text in texts]
        query_tokens = preprocess(doc_text)
        
        # Простая реализация BM25
        N = len(corpus)
        avgdl = sum(len(doc) for doc in corpus) / N if N > 0 else 0
        k1 = 1.5
        b = 0.75
        
        # Подсчет document frequency
        df = defaultdict(int)
        for document in corpus:
            frequencies = Counter(document)
            for word in frequencies:
                df[word] += 1
        
        # IDF
        idf = {}
        for word, freq in df.items():
            idf[word] = math.log(1 + (N - freq + 0.5) / (freq + 0.5))
        
        # BM25 scores
        bm25_scores = []
        for idx, document in enumerate(corpus):
            score = 0.0
            doc_freq = Counter(document)
            dl = len(document)
            
            for word in query_tokens:
                if word not in doc_freq:
                    continue
                idf_val = idf.get(word, 0)
                tf = doc_freq[word]
                denom = tf + k1 * (1 - b + b * dl / avgdl)
                score += idf_val * (tf * (k1 + 1)) / denom
            
            bm25_scores.append(score)
        
        bm25_ranked = sorted(range(len(bm25_scores)), key=lambda i: bm25_scores[i], reverse=True)
        
        # 3. Vector рейтинг
        try:
            model = SentenceTransformer('BAAI/bge-m3')
            query_embedding = model.encode(doc_text).tolist()
        except Exception as e:
            print(f"Ошибка генерации эмбеддинга: {e}")
            return []
        
        # Косинусное сходство
        vector_scores = []
        for emb in embeddings:
            if len(emb) == len(query_embedding):
                dot_product = sum(a * b for a, b in zip(emb, query_embedding))
                norm_a = math.sqrt(sum(a * a for a in emb))
                norm_b = math.sqrt(sum(b * b for b in query_embedding))
                if norm_a > 0 and norm_b > 0:
                    similarity = dot_product / (norm_a * norm_b)
                else:
                    similarity = 0
                vector_scores.append(similarity)
            else:
                vector_scores.append(0)
        
        vector_ranked = sorted(range(len(vector_scores)), key=lambda i: vector_scores[i], reverse=True)
        
        # 4. RRF (Reciprocal Rank Fusion)
        def rrf(idx, ranked_list):
            try:
                return ranked_list.index(idx)
            except ValueError:
                return len(ranked_list)
        
        c = 60  # RRF константа
        rrf_scores = {}
        for i in range(len(ids)):
            rv = rrf(i, vector_ranked)
            rf = rrf(i, bm25_ranked)
            rrf_scores[i] = 1/(c + rv) + 1/(c + rf)
        
        # 5. Выбираем топ-k уникальных правил
        top_rrf = sorted(rrf_scores, key=rrf_scores.get, reverse=True)[:k*2]
        seen_rule_ids = set()
        results = []
        
        for idx in top_rrf:
            chunk = chunks[idx]
            rule_id = chunk['rule_id']
            if rule_id in seen_rule_ids:
                continue
            seen_rule_ids.add(rule_id)
            
            # Получаем полный текст правила
            rule_query = f"""
                SELECT rule_text FROM rules WHERE rule_id = {rule_id} LIMIT 1
            """
            rule_df = pd.read_sql(rule_query, engine)
            if not rule_df.empty:
                rule_text = rule_df.iloc[0]['rule_text']
                if rule_text:
                    results.append(rule_text)
            
            if len(results) >= k:
                break
        
        print(f"RRF ретрив: найдено {len(results)} статей")
        return results
        
    except Exception as e:
        print(f"Ошибка выполнения RRF ретрива: {e}")
        return []

def evaluate_retrieval_metrics_rrf(filtered_docs, filtered_proof_texts, k=5):
    """
    Оценивает качество RRF ретрива с правильным подсчетом пересечений.
    """
    results = []
    
    for i, (doc, reference_texts_list) in enumerate(zip(filtered_docs, filtered_proof_texts)):
        print(f"\n--- Оценка документа {i+1} (RRF ретрив) ---")
        
        # 1. Получаем статьи через RRF ретрив
        retrieved_articles = retrieve_articles_rrf(doc, k=k, engine=engine)
        
        # Проверяем, что retrieved_articles это список
        if not isinstance(retrieved_articles, list):
            print(f"Ошибка: retrieved_articles не является списком, тип: {type(retrieved_articles)}")
            continue
            
        if not retrieved_articles:
            print("Не найдено статей через RRF ретрив, пропускаем документ")
            continue
        
        # 2. Подготавливаем эталонные тексты
        reference_text_combined = " ".join([text for text in reference_texts_list if text])
        reference_words = preprocess_text_for_metrics(reference_text_combined)
        
        # 3. Подготавливаем retrieved тексты
        retrieved_text_combined = " ".join(retrieved_articles)
        retrieved_words = preprocess_text_for_metrics(retrieved_text_combined)
        
        # 4. Подготавливаем отдельные чанки для Average Relevance
        chunks_words = [preprocess_text_for_metrics(article) for article in retrieved_articles]
        
        # 5. Рассчитываем метрики
        avg_relevance = calculate_average_relevance(chunks_words, reference_words)
        context_completeness = calculate_context_completeness(retrieved_words, reference_words)
        context_relevance = calculate_context_relevance(retrieved_words, reference_words)
        
        # 6. Правильно считаем пересечения
        intersection_count = 0
        found_reference_indices = set()
        
        for retrieved_article in retrieved_articles:
            retrieved_article_words = preprocess_text_for_metrics(retrieved_article)
            
            for ref_idx, reference_text in enumerate(reference_texts_list):
                if reference_text and ref_idx not in found_reference_indices:
                    reference_text_words = preprocess_text_for_metrics(reference_text)
                    if len(retrieved_article_words & reference_text_words) / len(reference_text_words) > 0.3:
                        intersection_count += 1
                        found_reference_indices.add(ref_idx)
                        break
        
        # 7. Сохраняем результаты
        doc_result = {
            'doc_id': i+1,
            'retrieved_articles_count': len(retrieved_articles),
            'reference_articles_count': len([text for text in reference_texts_list if text]),
            'intersection_count': intersection_count,
            'reference_words_count': len(reference_words),
            'retrieved_words_count': len(retrieved_words),
            'average_relevance': avg_relevance,
            'context_completeness': context_completeness,
            'context_relevance': context_relevance
        }
        
        results.append(doc_result)
        
        print(f"Эталонных статей: {doc_result['reference_articles_count']}")
        print(f"Статей найдено: {len(retrieved_articles)}")
        print(f"Пересечений найдено: {intersection_count}")
        print(f"Average Relevance: {avg_relevance:.4f}")
        print(f"Context Completeness: {context_completeness:.4f}")
        print(f"Context Relevance: {context_relevance:.4f}")
    
    return results

# Запускаем оценку RRF ретрива
print("\n" + "="*50)
print("ОЦЕНКА КАЧЕСТВА RRF РЕТРИВА")
print("="*50)

metrics_results_rrf = evaluate_retrieval_metrics_rrf(filtered_docs, filtered_proof_texts, k=5)
df_metrics_rrf = pd.DataFrame(metrics_results_rrf)

# Сравнение обычного и RRF ретрива
if len(metrics_results_with_intersections) > 0 and len(metrics_results_rrf) > 0:
    print(f"\n" + "="*50)
    print("СРАВНЕНИЕ ОБЫЧНОГО И RRF РЕТРИВА")
    print("="*50)
    
    # Обычный ретрив
    avg_rel_normal = sum(r['average_relevance'] for r in metrics_results_with_intersections) / len(metrics_results_with_intersections)
    cc_normal = sum(r['context_completeness'] for r in metrics_results_with_intersections) / len(metrics_results_with_intersections)
    cr_normal = sum(r['context_relevance'] for r in metrics_results_with_intersections) / len(metrics_results_with_intersections)
    
    # RRF ретрив
    avg_rel_rrf = sum(r['average_relevance'] for r in metrics_results_rrf) / len(metrics_results_rrf)
    cc_rrf = sum(r['context_completeness'] for r in metrics_results_rrf) / len(metrics_results_rrf)
    cr_rrf = sum(r['context_relevance'] for r in metrics_results_rrf) / len(metrics_results_rrf)
    
    print(f"Обычный ретрив:")
    print(f"  - Average Relevance: {avg_rel_normal:.4f}")
    print(f"  - Context Completeness: {cc_normal:.4f}")
    print(f"  - Context Relevance: {cr_normal:.4f}")
    
    print(f"\nRRF ретрив:")
    print(f"  - Average Relevance: {avg_rel_rrf:.4f}")
    print(f"  - Context Completeness: {cc_rrf:.4f}")
    print(f"  - Context Relevance: {cr_rrf:.4f}")
    
    print(f"\nУлучшение:")
    print(f"  - Average Relevance: {((avg_rel_rrf - avg_rel_normal) / avg_rel_normal * 100):+.1f}%")
    print(f"  - Context Completeness: {((cc_rrf - cc_normal) / cc_normal * 100):+.1f}%")
    print(f"  - Context Relevance: {((cr_rrf - cr_normal) / cr_normal * 100):+.1f}%")
```
